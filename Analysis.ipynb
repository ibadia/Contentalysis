So we are going to analyze the data we scraped from Mangobaaz.com, So our basic task over here is to first load the data into the pandas notebook and do some cool analysis on it.

```python
from collections import Counter
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk import bigrams, trigrams
import nltk
data0=pd.read_csv("Mangobaaz1.csv")
data1=pd.read_csv("Mangobaaz2.csv")
data2=pd.read_csv("Mangobaaz3.csv")

data=pd.concat([data0,data1,data2])

data['Article']=data['Article'].apply(lambda x: x.replace('"',''))
data['Article']=data['Article'].apply(lambda x: x.replace("'",""))
data['Article']=data['Article'].apply(lambda x : x.replace('“','').replace('”',''))
data['Article']=data['Article'].apply(lambda x : x.replace('.',''))
data['Article']=data['Article'].apply(lambda x : x.replace('’',''))



data['Headline']=data['Headline'].apply(lambda x: x.replace('"',''))
data['Headline']=data['Headline'].apply(lambda x: x.replace("'",""))
data['Headline']=data['Headline'].apply(lambda x : x.replace('“','').replace('”',''))
data['Headline']=data['Headline'].apply(lambda x: x.replace(".",""))
data['Headline']=data['Headline'].apply(lambda x: x.replace("’",""))


```

```python
tokenized_words_headlines=[nltk.word_tokenize(x['Headline']) for i,x in data.iterrows()]
tokenized_words_articles=[nltk.word_tokenize(x['Article']) for i,x in data.iterrows()]
```

```python
all_text=' '.join(list(data.Article))
all_text=all_text.replace('will','')
all_text=all_text.replace('pakistani',' ')
all_text=all_text.replace('pakistan',' ')
all_text=all_text.replace('new',' ')
all_text=all_text.replace('heres','')
```

```python
# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0).generate(all_text.lower())
 
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
