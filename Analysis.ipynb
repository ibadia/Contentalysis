```python
from collections import Counter
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
data=pd.read_csv("mangobaazTesting.csv")
data.head()
```

```python
article_word_frequency=Counter(' '.join(data.Headline).replace('\n',' ').split(' '))
headline_word_freuency=Counter(' '.join(data.Article).replace('\n',' ').split(' '))
```

```python
del article_word_frequency['']
article_word_frequency
```

```python
# Create the wordcloud object
wordcloud = WordCloud(width=480, height=480, margin=0).generate_from_frequencies(article_word_frequency)
 
# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
